```python
%matplotlib inline
import scipy as sp
import numpy as np
import math
from scipy import linalg
from matplotlib import pylab as plt
```

```python
def f (x):
    return np.sin(x/5)*np.exp(x/10) + 5*np.exp(-x/2)

x = np.arange(1, 15.1, 0.1)
y = f(x)
#print x
#print y
```

```python
#Многочлен степени 1

a1 = np.array([[1,1], [1,15]])
#print a1
b1 = np.array([f(1.),f(15.)])
#print b1

w1 = sp.linalg.solve(a1,b1)
#print w1

y1 = w1[0] + w1[1] * x

plt.plot(x, y)
plt.plot(x, y1)
```

```python
#Многочлен степени 2

a2 = np.array([[1,1,1], [1, 8, 8*8], [1,15,15*15]])
#print a2
b2 = np.array([f(1.),f(8.),f(15.)])
#print b2

w2 = sp.linalg.solve(a2,b2)
#print w2

y2 = w2[0] + w2[1] * x + w2[2]*x*x

plt.plot(x, y)
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
#Многочлен степени 3

a3 = np.array([[1,1,1, 1], [1, 4, 4**2, 4**3], [1, 10, 10**2, 10**3], [1,15,15**2, 15**3]])
#print a3
b3 = np.array([f(1.),f(4.),f(10.),f(15.)])
#print b3

w3 = sp.linalg.solve(a3,b3)
print w3

y3 = w3[0] + w3[1] * x + w3[2]*x**2 + w3[3]*x**3

plt.plot(x, y)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
```
